```python
from google.cloud import storage
import pandas as pd

# Define your storage client
storage_client = storage.Client('fa23-i535-savinn-spotify-songs')

# Define your bucket name
bucket_name = 'bigdata_spotify'

# Define your file name
file_name = 'universal_top_spotify_songs.csv'

# Create a bucket object
bucket = storage_client.get_bucket(bucket_name)

# Create a blob object
blob = bucket.blob(file_name)

# Download the contents of the blob as a string and then parse it using pandas
blob.download_to_filename('/content/' + file_name)
data = pd.read_csv('/content/' + file_name)
```

```python
data['country'] = data['country'].fillna('Unknown')
```

```python
data= data.dropna()
```

```python
data['album_release_date'] = pd.to_datetime(data['album_release_date'])
data['snapshot_date'] = pd.to_datetime(data['snapshot_date'])
```

```python
from scipy import stats
import numpy as np

# Assuming 'numerical_column' is your data of interest
z_scores = np.abs(stats.zscore(data['mode']))
data = data[(z_scores < 3)]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = data.corr()
plt.figure(figsize=(17, 17))
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```

```python
Q1 = data['popularity'].quantile(0.25)
Q3 = data['popularity'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
data_filtered = data[(data['popularity'] >= lower_bound) & (data['popularity'] <= upper_bound)]
```

```python
# Select only the numerical columns from the DataFrame
numerical_data = data.select_dtypes(include=['int64', 'float64'])

# Set up the matplotlib figure
plt.figure(figsize=(20, 15))

# Create a histogram for each numerical column
for index, column in enumerate(numerical_data.columns):
    plt.subplot(5, 5, index + 1)  # Adjust the layout parameters as needed
    sns.histplot(numerical_data[column], kde=False, bins=15)
    plt.title(column)
    plt.tight_layout()

plt.show()
```
